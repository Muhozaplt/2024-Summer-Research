#### Definitions:

```python
# titles definition for setting xlim and ylim, and axis titles of the plots
def titles(xlim1, xlim2, ylim1, ylim2, xtitle, ytitle):
    plt.xlim(xlim1, xlim2)
    plt.ylim(ylim1, ylim2)
    plt.ylabel(ytitle)
    plt.xlabel(xtitle)
    
```

```python
# definition for UVJ diagram axis limits and star-forming, Quiescent, dusty division lines
def uvj_dec(x1, x2, y1, y2):
    plt.xlim(x1, x2)# x-axis limits
    plt.ylim(y1, y2)# y-axis limits
    
    plt.plot([-100000, 0.85, 1.6, 1.6], [1.3, 1.3, 1.95, 100000], c='black', alpha=0.5) #separation lines on the uvj diagram
    plt.plot([1.2, 1.2], [0, 1.6], linestyle='--', c= 'black', alpha=0.5)
    # axis titles
    plt.ylabel('Restframe U - V   [Mag]')
    plt.xlabel('Restframe V - J   [Mag]')
    plt.title("COSMOS legacy field UVJ diagram")
    # Quiescent, star forming and dusty region titles
    plt.text(0.05, 0.95, 'Quiescent', ha='left', va='top', transform=plt.gca().transAxes, c='red')
    plt.text(0.005, 0, 'Star Forming', ha='left', va='bottom', transform=plt.gca().transAxes, c='blue')
    plt.text(0.89, 0, 'Dusty', ha='left', va='bottom', transform=plt.gca().transAxes, c='orange')
```

# CIGALE -COSMOS field data generated
For analysis of the ZFOURGE survey [(Straatman et al.)](https://iopscience.iop.org/article/10.3847/0004-637X/830/1/51), this research used CIGALE [‘CIGALE – Code Investigating GALaxy Emission’](https://cigale.lam.fr/) method which compares modelled galaxy spectral energy distributions (SEDs) to observed ZFOURGE data. CIGALE then generated AGN luminosity data of each galaxy which was used for further analysis seeking for contamination on the UVJ diagram.

```python
import numpy as np
import pandas as pd

cigale_output= np.genfromtxt("cigale_output.txt", names=True) #loading the CIGALE data file
cigale_output=pd.DataFrame(cigale_output) # creatng a pandas dataframe
cigale_output # print output
```

```python
# loading the COSMOS fiels legacy data form cosmos.v1.3.6.fout which contain stellar mass ("lmass"), star formation rate and redshift ("z") 
cosmos_data1= np.genfromtxt("cosmos.v1.3.6.fout", names=True) 
cosmos_data1= pd.DataFrame(cosmos_data1)
cosmos_data1 # print the dataframe table
```

```python
# observing the spread of agn luminosity data
cigale_output['bayesagnluminosity'].describe() 
```

```python
# Plotting the COSMOS data 
import matplotlib.pyplot as plt
plt.scatter(cosmos_data1['z'], cosmos_data1['lmass'], s=3, alpha=0.6) # stellar mass vs redshift 
titles(-0.1, 5, 4, 11.5, 'redshift', 'Stellar Mass (log[M/M$_{\odot}$])')
plt.title('Redshift vs stellar mass of the COSMOS legacy field')
```

```python
# Calculating the star-formation rate 
sfr_uv= np.genfromtxt('cosmos.v1.3.6.sfr.v0.4.cat', names=True) # loading star-formation data in .cat file
sfr_uv=pd.DataFrame(sfr_uv)

ssfr= sfr_uv['SFR_UVIR']/(10**cosmos_data1['lmass']) # caltulations of ssfr
sfr_uv['ssfr']= ssfr 
sfr_uv['log[ssfr]']= np.log10(sfr_uv['ssfr'])
sfr_uv
```

#### Rest-frame calculations (U-V, V-J)
For these claculations to convert from rest-frame fluxes to AB magnitudes, we used the form formula from [Straatman et al., ‘The FourStar Galaxy Evolution Survey (ZFOURGE)’.](https://iopscience.iop.org/article/10.3847/0004-637X/830/1/51):

$RF_{AB}= 25-2.5log_{10}(f_{rest})$

```python
cos_rest_data= np.genfromtxt("cosmos.v1.3.6.rest.v0.9.cat", names=True) # Loading COSMOS restframe data
cos_rest_data= pd.DataFrame(cos_rest_data)


# creating new columns for the Mag calcs
cos_rest_data['U_mag']=25-2.5*np.log10(cos_rest_data['f_U']) 
cos_rest_data['V_mag']=25-2.5*np.log10(cos_rest_data['f_V'])
cos_rest_data['J_mag']=25-2.5*np.log10(cos_rest_data['f_J'])

# calculating U-V and V-J
cos_rest_data['V-J'] = cos_rest_data['V_mag']-cos_rest_data['J_mag']
cos_rest_data['U-V'] = cos_rest_data['U_mag']-cos_rest_data['V_mag']

# adding star formation rate, redshift and stellar mass to the rest of the data

cos_rest_data['ssfr']= sfr_uv['log[ssfr]'] #star formation rate
cos_rest_data['z']= cosmos_data1['z'] #redshift 
cos_rest_data['lmass']= cosmos_data1['lmass'] # add stellar mass to the data
cos_rest_data
```

#### Merging the COSMOS rest frame data and Cigale-output in one dataframe


```python
merged_data = pd.merge(cos_rest_data, cigale_output, left_on='id', right_on='your_id', how='inner') # merges the restframe data and cigale output based on matching IDs in both files
merged_data
```

#### Splitting data by averages bins of redshifts

```python
def averages (data):
    df = data
# Specify the bin edges (you can adjust the bin size as needed)
    bin_edges = [1, 1.25, 1.5, 1.75, 2, 2.25, 2.5] # I created 6 bins of these intervals
# Create bins based on '['V-J']'  averaged
    df['Bins'] = pd.cut(df['z'], bins=bin_edges, labels=False, right=False)
# Group by the bins and calculate the average for each group
    result_df = df.groupby('Bins').mean()
    return result_df
```

### Errror bars calculations
In this research, we calculated errors by "Uncertainties and Error Propagation" approach

|Formulas|
---------

Magnitude:

$\delta M = \frac{2.5}{\ln(10).flux} \delta F$

Once the magnitude has been calculated, we propagate the error through V-J and U-V:

$Q=\sqrt{M_U^2+M_V^2}$ and do the same for V-J

```python
## Magnitude calculation using the formula above
M_U= (2.5)/(np.log(10)*merged_data['f_U'])* merged_data['e_U']
M_J= (2.5)/(np.log(10)*merged_data['f_J'])* merged_data['e_J']
M_V= (2.5)/(np.log(10)*merged_data['f_V'])* merged_data['e_V']

merged_data['err_uv']=np.sqrt((M_U)**2+(M_V)**2) # error  claculation of U-V
merged_data['err_vj']=np.sqrt((M_V)**2+(M_J)**2)# error  claculation of V-J
```

```python
# AGN host data

# Create a scatter plot with error bars
plt.errorbar(merged_data['V-J'][merged_data['bayesagnfracAGN']>0.5], merged_data['U-V'][merged_data['bayesagnfracAGN']>0.5], xerr=merged_data['err_vj'][merged_data['bayesagnfracAGN']>0.5], yerr=merged_data['err_uv'][merged_data['bayesagnfracAGN']>0.5], fmt='o', capsize=5, markersize=8, label='Data with Error Bars', alpha = 0.4)

# Customize the plot
uvj_dec(0, 2.5, 0, 2.5)
plt.title('Galaxies of >50% AGN luminosity')
plt.grid(True)
plt.show()
```
